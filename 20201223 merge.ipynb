```python
import pandas as pd  
import numpy as np  
import glob  
import sys
```

```python
#파일 Union  
all_data = pd.DataFrame()  
for f in glob.glob('C:/Users/nfri_def/Desktop/RPA/download/excel/abstract_*.xls'): #폴더경로 입력하세요. 
    df = pd.read_excel(f)
    all_data = all_data.append(df, ignore_index=True)
```

```python
#데이터갯수확인  
print(all_data.shape)
```

```python
#데이터 잘 들어오는지 확인  
all_data.head()
```

```python
#파일저장  
all_data.to_excel("C:/Users/nfri_def/Desktop/RPA/download/excel/merge.xls", header=False, index=False) #헤더없이 합친후, 헤더는 나중에 추가했습니다. 
```
